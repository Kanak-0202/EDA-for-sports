# Now analysing the bowling rate 

```python
no_of_wickets= df.groupby('bowler').agg({'total_run':'sum','bowler':'count','isWicketDelivery':'sum'})
no_of_wickets.rename(columns={'bowler':'no_of_balls'},inplace=True)
no_of_wickets
```

```python
no_of_wickets['bowlingRate'] = no_of_wickets['total_run']/no_of_wickets['no_of_balls']*6
no_of_wickets
```

## Finding top players according to bowling rate

```python
no_of_wickets.loc[no_of_wickets['no_of_balls'] >= 60, 'best_wicket_delivery'] =no_of_wickets['bowlingRate']
t= no_of_wickets.sort_values(by='best_wicket_delivery').head(10)
t
```
